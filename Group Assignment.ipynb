```python
fileHandler = open('C:/Users/User/Desktop/TXSA Assignment/Group Assignment Data/Text Corpus.txt')
text = fileHandler.read()
print(text)

sentenceTokens = text.split('\n') # split the sentence by new line
print("\nTotal number of sentences in this text corpus is",len(sentenceTokens),"\n")
for i in sentenceTokens:
    print(i)
    
len(sentenceTokens)
print(sentenceTokens)
```

```python
import nltk
cleanedTokens=[]
for i in sentenceTokens:
    newSentences = i.replace('<s>','').replace('</s>','') #remove the start and end symbol
    print(newSentences,"\n")
    cleanedTokens.append(newSentences)
```

```python
#1.3 Unsmoothed unigram
import nltk
from nltk.util import ngrams

unsmoothedUnigram=[]
for a in cleanedTokens:
    word = nltk.tokenize.word_tokenize(a) #word tokenisation
    listing = list(ngrams(word, n=1)) #perform unsmoothed unigram language model
    #unsmoothedUnigram.append(listing)
    unsmoothedUnigram.append(word)
print(listing)
print(unsmoothedUnigram)
```

```python
for i in unsmoothedUnigram:
    print(i)
    
print(unsmoothedUnigram)
```

```python
for i in unsmoothedUnigram: #vocab count per sentence
    fdist1 = nltk.FreqDist(i)
    print('Count per vocabs of sentence',unsmoothedUnigram.index(i)+1,':',fdist1.most_common())

uniquevocab = []
for i in unsmoothedUnigram:
    for j in i:
        if j not in uniquevocab:
            uniquevocab.append(j)
print('\nTotal number of unique vocab:',len(uniquevocab))

ano = sum(arr.count('a') for arr in unsmoothedUnigram) #sum of vocab count among all sentences
bno = sum(arr.count('b') for arr in unsmoothedUnigram)
cno = sum(arr.count('c') for arr in unsmoothedUnigram)
print('\nTotal number of a:',ano,'\nTotal number of b:',bno,'\nTotal number of c:',cno)
thesum = ano+bno+cno

#Unsmoothed unigram probability
usuapv = ano/thesum
usubpv = bno/thesum
usucpv = cno/thesum
print('\nProbability of a:',usuapv,'\nProbaility of b:',usubpv,'\nProbability of c:',usucpv)
```

```python
#1.3 Smoothed unigram probability
suapv = (ano + 1)/(thesum + len(uniquevocab) + 1)
subpv = (bno + 1)/(thesum + len(uniquevocab) + 1)
sucpv = (cno + 1)/(thesum + len(uniquevocab) + 1)
print('Probability of a:',suapv,'\nProbaility of b:',subpv,'\nProbability of c:',sucpv)
```

```python
#2.3 Unsmoothed bigram
from nltk.util import bigrams
from nltk.lm.preprocessing import pad_both_ends

unsmoothedBigram = []
for i in sentenceTokens:
    word2 = i.split(' ')
    listing2 = list(nltk.bigrams(word2))
    unsmoothedBigram.append(listing2)

for i in unsmoothedBigram:
    print(i)
    
print('\n',unsmoothedBigram)

print('\nBigram count:')

for i in unsmoothedBigram:
    fdist2 = nltk.FreqDist(i)
    print(fdist2.most_common())
    
uniquevocab2 = []
for i in unsmoothedBigram:
    for j in i:
        if j not in uniquevocab2:
            uniquevocab2.append(j)
            
print('\nTotal number of bigram set:',len(uniquevocab2))

print('\n',uniquevocab2)
```

```python
#4.1 Naïve Bayes Classifier from NLTK 

import nltk
import csv
countrow = poscount = negcount = neucount = 0
csvlist = []
with open('C:/Users/User/Desktop/TXSA Assignment/Group Assignment Data/Financial News Dataset.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
        csvlist.append(row)
        countrow+=1

for row in csvlist:
    if(row[0] == 'positive'):
        poscount +=1
    elif(row[0] == 'negative'):
        negcount +=1
    elif(row[0] == 'neutral'):
        neucount +=1

    
print("\nTotal positive news:", poscount, "\nTotal negative news:", negcount, "\nTotal neutral news:", neucount)
print("Total news:", poscount+negcount+neucount)
print(countrow) #print the count of row
```

```python
import random #random shuffle the list

random.shuffle(csvlist)

newslist=[]
bagofwords = []

for row in csvlist: #move to newslist according to the rows mentioned
    for r in range(0, len(row), 2):
        newslist.append(row[r])
    for r in range(1, len(row), 2):
        lm = nltk.tokenize.word_tokenize(row[r].lower()) #tokenize and lowercase
        newslist.append(lm)
        
print(newslist)

for words in newslist:
    if(words != 'positive' and words != 'negative' and words != 'neutral'):
        for w in words:       
            bagofwords.append(w)
```

```python
print(bagofwords)
```

```python
bagofwords = nltk.FreqDist(bagofwords)
print(bagofwords.most_common(15)) # print the most common word found in the wordset
print(bagofwords['dropped']) # to print the occurence of word dropped 
```

```python
documents = []
print(len(newslist))
for n in range(0,len(newslist),2):
    doc = (newslist[n+1], newslist[n])
    documents.append(doc)    
print(documents)
```

```python
keywords = list(bagofwords)[:3000] # to select first 3000 vocabs
print(keywords)
```

```python
def matchkeywords(document):
    words = set(document)
    features = {}
    for k in keywords:
        features[k] = (k in words) # if the word matches with in the top 3k words, return true
    return features
```

```python
import nltk
import collections

featureset = [(matchkeywords(news), sentiment)  for (news, sentiment) in documents]

trainingset = featureset[:4000] # training with 4000 data row
testingset = featureset[4000:] # the remaining goes to test

classifier = nltk.NaiveBayesClassifier.train(trainingset) # train the set
print(nltk.classify.accuracy(classifier,testingset)*100)
classifier.show_most_informative_features(15) # show 15 most informative words
```

```python
from nltk.classify import NaiveBayesClassifier
classifier2 = NaiveBayesClassifier.train(trainingset)
refsets = collections.defaultdict(set)
testsets = collections.defaultdict(set)

for i, (feats, label) in enumerate(testingset):
    refsets[label].add(i)
    observed = classifier2.classify(feats)
    testsets[observed].add(i)

print("The accuracy is: ", nltk.classify.accuracy(classifier2, testingset) * 100, "%")
print ('\npos precision:', nltk.scores.precision(refsets['positive'], testsets['positive']))
print ('pos recall:', nltk.scores.recall(refsets['positive'], testsets['positive']))
print ('pos F-measure:', nltk.scores.f_measure(refsets['positive'], testsets['positive']))
print ('\nneg precision:', nltk.scores.precision(refsets['negative'], testsets['negative']))
print ('neg recall:', nltk.scores.recall(refsets['negative'], testsets['negative']))
print ('neg F-measure:', nltk.scores.f_measure(refsets['negative'], testsets['negative']))
```
